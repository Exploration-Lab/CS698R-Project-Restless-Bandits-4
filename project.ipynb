```python
import numpy as np
import gym
import matplotlib.pyplot as plt
```

```python
env1 = gym.make('my_env:restlessbandits-v0')
env2 = gym.make('my_env:restlessbandits-v0')
env2.__init__(high = True)
env3 = gym.make('my_env:restlessbandits-v0')
env3.__init__(trend = False)
env4 = gym.make('my_env:restlessbandits-v0')
env4.__init__(trend = False, high = True)
```

```python
rew1 = np.zeros((10, 200, 4))
rew2 = np.zeros((10, 200, 4))
rew3 = np.zeros((10, 200, 4))
rew4 = np.zeros((10, 200, 4))
for k in range(10):
    rng = np.random.RandomState(seed = 1220+k)
    for j in range(4):
        for i in range(200):
            _,rew1[k][i][j],_,_ = env1.step(j, rng=rng)
            _,rew2[k][i][j],_,_ = env2.step(j, rng=rng)
            _,rew3[k][i][j],_,_ = env3.step(j, rng=rng)
            _,rew4[k][i][j],_,_ = env4.step(j, rng=rng)
        env1.reset()
        env2.reset()
        env3.reset()
        env4.reset()
plt.figure(figsize=(15, 10))
plt.subplot(2,2,2)
plt.plot(np.average(rew1, axis=0), label = 'Trend-Low')
plt.ylim([-150, 150])
plt.legend()
plt.subplot(2,2,4)
plt.plot(np.average(rew2, axis=0), label = 'Trend-High')
plt.ylim([-150, 150])
plt.legend()
plt.subplot(2,2,1)
plt.plot(np.average(rew3, axis=0), label = 'NoTrend-Low')
plt.ylim([-150, 150])
plt.legend()
plt.subplot(2,2,3)
plt.plot(np.average(rew4, axis=0), label = 'NoTrend-High')
plt.ylim([-150, 150])
plt.legend()
plt.show()
```

```python
from epsgreedy import epsilongreedy
from ucb import ucbfunc
from softmax import softmaxfunc
```

```python
rw11 = np.zeros((50, 200))
rw12 = np.zeros((50, 200))
rw13 = np.zeros((50, 200))
rw14 = np.zeros((50, 200))
rw15 = np.zeros((50, 200))
rw16 = np.zeros((50, 200))

re11 = np.zeros((50, 200))
re12 = np.zeros((50, 200))
re13 = np.zeros((50, 200))
re14 = np.zeros((50, 200))
re15 = np.zeros((50, 200))
re16 = np.zeros((50, 200))

oa11 = np.zeros((50, 200))
oa12 = np.zeros((50, 200))
oa13 = np.zeros((50, 200))
oa14 = np.zeros((50, 200))
oa15 = np.zeros((50, 200))
oa16 = np.zeros((50, 200))

for i in range(50):
    seed = 3333 + i
    rng = np.random.RandomState(seed)
    oa11[i,:], re11[i, :], rw11[i, :],_ ,_ = epsilongreedy(env1, maxep = 200, epsilon = 1, rng = rng)
    oa12[i,:], re12[i, :], rw12[i, :],_ ,_ = epsilongreedy(env1, maxep = 200, epsilon = 0, rng = rng)
    oa13[i,:], re13[i, :], rw13[i, :],_ ,_ = epsilongreedy(env1, maxep = 200, epsilon = 0.1, rng = rng)
    oa14[i,:], re14[i, :], rw14[i, :],_ ,_ = epsilongreedy(env1, maxep = 200, epsilon = 1, decay = 0.01, isdecaylinear = True, rng = rng)
    oa15[i,:], re15[i, :], rw15[i, :],_ = softmaxfunc(env1, maxep = 200, rng = rng)
    oa16[i,:], re16[i, :], rw16[i, :],_ = ucbfunc(env1, maxep = 200, rng = rng)
```
